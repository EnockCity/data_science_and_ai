```python
# simple line plot
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, pop)
plt.show()
```

```python
# simple line plot with labels
import matplotlib.pyplot as plt

# Sample data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Creating plot
plt.plot(x, y, marker='o', linestyle='-', color='b')
# Add titles and labels for the axes
plt.title("Sample data line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show plot
plt.show()
```
