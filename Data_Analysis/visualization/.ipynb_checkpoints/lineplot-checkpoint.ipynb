Line plot
A line plot is useful for visualizing data points over time

```python
# simple line plot
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, pop)
plt.show()
```

```python
# simple line plot with labels
import matplotlib.pyplot as plt

# Sample data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Creating plot
plt.plot(x, y, marker='o', linestyle='-', color='b')
# Add titles and labels for the axes
plt.title("Sample data line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Creating plot
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Setting  custom x-ticks
plt.xticks([0, 1, 2, 3, 4, 5])

# Add titles and labels
plt.title('Sample of Custom X-Ticks')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Create plot
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Set custom x-ticks and labels
plt.xticks([0, 1, 2, 3, 4, 5], ['zero', 'one', 'two', 'three', 'four', 'five'], rotation=45, fontsize='small')

# Set custom y-ticks and labels
plt.yticks([0, 5, 10, 15, 20, 25], ['a', 'b', 'c', 'd', 'e', 'f'], rotation=45, fontsize='small')

# Add titles and labels
plt.title('Custom Tick Appearance')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Create plot
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Set major ticks
plt.xticks([0, 1, 2, 3, 4, 5])
plt.yticks([0, 5, 10, 15, 20, 25])

# Add minor ticks
plt.minorticks_on()
plt.grid(which='both')  # Add grid lines for both major and minor ticks
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

# Add titles and labels
plt.title('Sample of Major and Minor Ticks')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Create plot
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Add text annotation
plt.text(2, 10, 'Text Annotation', fontsize=12, color='red')

# Add titles and labels
plt.title('Example of Text Annotation')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Create plot
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Add text annotation with advanced customization
plt.text(2, 10, 'Custom Text Annotation',
         fontsize=14, color='blue', fontstyle='italic', fontweight='bold',
         rotation=45, backgroundcolor='yellow',
         bbox=dict(facecolor='red', alpha=0.5))

# Add titles and labels
plt.title('Sample of Advanced Text Annotation')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Create plot
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Annotate each point with its coordinates
for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})', fontsize=12, ha='right')

# Add titles and labels
plt.title('Annotating Data Points')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Show plot
plt.show()
```
